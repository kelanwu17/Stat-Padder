```python

from nba_api.stats.static import teams, players
from nba_api.stats.endpoints import playerawards, commonplayerinfo,franchisehistory, TeamYearByYearStats
import pandas as pd
import json
nba_teams = teams.get_teams()
rockets = [team for team in nba_teams if team['abbreviation'] == 'HOU'][0]
rockets_id = rockets['id']

print(rockets)
print(f'rockets_id: {rockets_id}')


```

```python
nba_players = players.get_players()
player_name = 'Jordan Poole'
player = [player for player in nba_players if player['full_name'] == player_name][0]
player_id = player['id']
print(player)

player_awards = playerawards.PlayerAwards(player_id = player_id)
player_json = player_awards.get_normalized_json()
player_json_loads = json.loads(player_json)
print(json.dumps(player_json_loads, indent = 4))
all_defense_team = 0
all_nba = 0
fmvp = 0
mvp = 0
roty = 0


```

```python
for i in range(0,len(player_json_loads["PlayerAwards"])):
    if (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "All-Defensive Team"):
        all_defense_team += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "All-NBA"):
        all_nba += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Finals Most Valuable Player"):
        fmvp += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Most Valuable Player"):
        mvp += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Rookie of the Year"):
        roty += 1
print(all_defense_team, "x All_Defensive Team")
print(all_nba, "x All_NBA")
print(fmvp, "x Finals MVP")
print(mvp, "x MVP")
print(roty, "x Rookie of the Year")
```

```python
player_info = json.loads((commonplayerinfo.CommonPlayerInfo(player_id = player_id)).get_normalized_json())
player_team_id = player_info["CommonPlayerInfo"][0]["TEAM_ID"]
print(player_team_id)
print(player_info)
```
