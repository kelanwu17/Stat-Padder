```python
player_input = 'jalen green'
nba_players = json.loads(commonallplayers.CommonAllPlayers().get_normalized_json())
players = nba_players['CommonAllPlayers'][0]
player_id = 0;
player_name = ""

for i in range(0, len(nba_players['CommonAllPlayers'])):
    if nba_players['CommonAllPlayers'][i]['DISPLAY_FIRST_LAST'].lower() == player_input.lower():
        player_id = nba_players['CommonAllPlayers'][i]['PERSON_ID']
        player_name = nba_players['CommonAllPlayers'][i]['DISPLAY_FIRST_LAST']
        break
stats = json.loads(playercareerstats.PlayerCareerStats(player_id = player_id).get_normalized_json())
careerpts = 0
careergames = 0
careerptsavg = 0
careerassists = 0
careerrebounds = 0
for i in range(0, len(stats['SeasonTotalsRegularSeason'])):
    if stats['SeasonTotalsRegularSeason'][i]['GP'] > 0:
        careergames = careergames + stats['SeasonTotalsRegularSeason'][i]['GP']
    if stats['SeasonTotalsRegularSeason'][i]['PTS'] > 0:
        careerpts = careerpts + stats['SeasonTotalsRegularSeason'][i]['PTS']
    if stats['SeasonTotalsRegularSeason'][i]['REB'] > 0:
        careerrebounds = careerrebounds + stats['SeasonTotalsRegularSeason'][i]['REB']
    if stats['SeasonTotalsRegularSeason'][i]['AST'] > 0:
        careerassists = careerassists + stats['SeasonTotalsRegularSeason'][i]['AST']
print(stats['SeasonTotalsRegularSeason'])
print("Games Played ", careergames)
print("TOTAL POINTS ", careerpts)
careerptsavg = careerpts/careergames
print("CAREER PTS AVG ", round(careerptsavg, 2))
print("Career Assists ", round(careerassists/careergames,2))
print("Career Reb ", round(careerrebounds/careergames,2))




```

```python
player_awards = playerawards.PlayerAwards(player_id = player_id)
player_json = player_awards.get_normalized_json()
player_json_loads = json.loads(player_json)
print(json.dumps(player_json_loads, indent = 4))
all_defense_team = 0
all_nba = 0
fmvp = 0
mvp = 0
roty = 0
for i in range(0,len(player_json_loads["PlayerAwards"])):
    if (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "All-Defensive Team"):
        all_defense_team += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "All-NBA"):
        all_nba += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Finals Most Valuable Player"):
        fmvp += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Most Valuable Player"):
        mvp += 1
    elif (player_json_loads["PlayerAwards"][i]['DESCRIPTION'] == "NBA Rookie of the Year"):
        roty += 1
print(all_defense_team, "x All_Defensive Team")
print(all_nba, "x All_NBA")
print(fmvp, "x Finals MVP")
print(mvp, "x MVP")
print(roty, "x Rookie of the Year")
```
